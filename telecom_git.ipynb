```python
def days_before_payment(row):
    if row['Type'] == 'Month-to-month':
        return round(row['total_days'] % (365.25/12))
    elif row['Type'] == 'One year':
        return round(row['total_days'] % 365.25)
    return round(row['total_days'] % 730.5)

df['days_before_payment'] = df.apply(days_before_payment, axis=1)
df.head(2)
```

```python
INTERVAL_COLS = ['monthly_charges', 'total_charges', 'days_before_payment', 'total_days']
corr_table = df.phik_matrix(interval_cols=INTERVAL_COLS)
```

```python
def proba_plot(data, column):
    array0 = []
    array1 = []
    unique_list = list(data[column].unique())
    for i in range(len(unique_list)):
        all_unique = data.loc[data[column] == unique_list[i], column].count()

        target_1 = data.loc[(data[column] == unique_list[i]) & (data['target'] == 1), column].count()
        proba1 = np.round((target_1 / all_unique), 2)
        target_0 = data.loc[(data[column] == unique_list[i]) & (data['target'] == 0), column].count()
        proba0 = np.round((target_0 / all_unique), 2)

        array0.append([unique_list[i], proba0, '–ù–µ —É—à—ë–ª'])
        array1.append([unique_list[i], proba1, '–£—à—ë–ª'])

    array = np.concatenate([array0, array1], axis=0)
    df = pd.DataFrame(data=array, columns = ['–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ '+column, '–î–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤', 'Hue'])
    df['–î–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤'] = df['–î–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤'].astype(float) # –û–±—ä—è—Å–Ω–∏—Ç–µ –ø–æ–∂–∞–ª—É–π—Å—Ç–∞ –ø–æ—á–µ–º—É –ø–æ—Å–ª–µ concatenate –¥–∞–Ω–Ω—ã–µ –ø–µ—Ä–µ—Ö–æ–¥—è—Ç —Å float –Ω–∞ object. –ü—Ä–∏—à–ª–æ—Å—å –ø–µ—Ä–µ–≤–æ–¥–∏—Ç—å, –æ—à–∏–±–∫—É –Ω–∞–∫ –∏ –Ω–µ –ø–æ–Ω—è–ª...
    display(df)
    
    sns.barplot(data = df, x = '–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ '+column, y = '–î–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤', hue = 'Hue')
    plt.show()
    
    print()
    print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è Phik:', np.round(corr_table.loc[column]['target'], 2))
    print('-----------------------------------------------------------------------------------------------')
    print()
    
    return None
```

```python
# –§—É–Ω–∫—Ü–∏—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –º–æ–¥–µ–ª–∏
def plot_feature_importance(feature_names, feature_scores, top_feats=20, title='Importance', x_label='Importance', 
                            y_label='Features'):
    tuples = [(name, round(score, 3)) for name, score in zip(feature_names, feature_scores)]
    tuples = sorted(tuples, key=lambda x: x[1])[-top_feats:]
    
    labels, values = zip(*tuples)
    _, ax = plt.subplots(1, 1, figsize=(10, 8))
    ylocs = np.arange(len(values))
    
    ax.barh(ylocs, values, align='center', height=0.4)
    for x, y in zip(values, ylocs):
        ax.text(x + 1, y, x, va='center')
        
    ax.set_yticks(ylocs)
    ax.set_yticklabels(labels)
    xlim = (0, max(values) * 1.1)
    ax.set_xlim(xlim)
    ylim = (-1, len(values))
    ax.set_ylim(ylim)
    
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.grid(True)
```

```python
importances = cbc_gs.best_estimator_.feature_importances_
plot_feature_importance(X_train, importances, top_feats=X_train.shape[1], title='CatBoost feature importance', 
                        x_label='Importance')
```
